```python
# 1. Request 取回之後該怎麼取出資料，資料型態是什麼？
import requests 
r = requests.get('https://www.dcard.tw/f') 
response = r.text 
print(response)

#資料型態是字串
print('資料型態=',type(response))
```

```python
# 2.為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？
# 字串經由BeautifulSoup處理後，會產生BeautifulSoup物件，此物件中包含了整個HTML文件的結構樹。方便我們從結構樹中取出想要的資料。
import requests 
from bs4 import BeautifulSoup

r = requests.get('https://www.dcard.tw/f') 
response = r.text 
soup = BeautifulSoup(response, "html5lib")
print(soup)

#資料型態是bs4.BeautifulSoup
print('資料型態=',type(soup))
```

```python
# 3.觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？
import requests 
r = requests.get('https://www.zhihu.com/explore') #上方cell已經爬過Dcard了，接下來爬知乎跟解析它。
response = r.text 
print(response)
print('資料型態=',type(response))

from bs4 import BeautifulSoup
r = requests.get('https://www.zhihu.com/explore') 
bs = r.text 
soup = BeautifulSoup(bs, "html5lib")
print(soup)
print('資料型態=',type(soup))
```

```python
#上方作答中，response出現了400 Bad Request。但直接用瀏覽器打開網誌，頁面是正常顯示的。
#也就是說發Request時有錯，原因是知乎的伺服器有做檢查的機制，會透過 Headers檢查「發送方」是否合法。
#所以發Request時要給Header。

import requests
from bs4 import BeautifulSoup

headers = {'user-agent': 'my-app/0.0.1'}
r = requests.get('https://www.zhihu.com/explore',headers=headers)
response = r.text
# print(response)

soup = BeautifulSoup(response, "html5lib")
print(soup)
```
