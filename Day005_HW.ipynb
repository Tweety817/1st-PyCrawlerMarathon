```python
# 1.這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
import json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

datas = json.loads(response)
print('這個 API 一次會回傳幾筆資料？ A:',len(datas))

keyList=[]
for key in datas[0].keys():
    keyList.append(key)
print('每一筆資料包含哪些欄位？',keyList)
```

```python
# 2.取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

import requests
import json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

datas = json.loads(response)
for data in datas:
    title = data['title']
    createdTime = data['createdAt']    
    commentCount = data['commentCount']    
    likeCount = data['likeCount']  
    print('標題=',title,'  貼文時間=',createdTime,'  留言人數=',commentCount,'  按讚人數=',likeCount)
       
```

```python
#3.計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
import requests
import json

def countPostInfo(boolen):
    r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular='+ boolen)
    response = r.text
    datas = json.loads(response)
    postCount = len(datas)
    totalCommentCount =0
    totalLikeCount =0
    for data in datas:  
        totalCommentCount+=data['commentCount']
        totalLikeCount+=data['likeCount']  
        
    if(boolen =="true"):
        s = "熱門"
    else:
        s = "非熱門"
    print(s+'文章的平均留言人數= ',totalCommentCount/postCount)
    print(s+'文章的平均按讚人數= ',totalLikeCount/postCount)
    
countPostInfo("true")
countPostInfo("false")
```
